```
edge_file = open("PP.txt", "r")
edge_list = []
count=0
test_edges=[]
my_network = nx.DiGraph()

for line in edge_file:
  x, y, z = map(int, line.split())
  if(np.random.random_sample()<0.9):
    if(count<1000):
      test_edges.append((x,y))
      count=count+1
  edge_list.append((x, y))

my_network.add_edges_from(edge_list)
graph = my_network
print(graph.number_of_nodes())
print(graph.number_of_edges())

for i in test_edges:
  graph.remove_edge(test_edges[i])
  
print(len(test_edges))
print(graph.number_of_nodes())
print(graph.number_of_edges())

```
